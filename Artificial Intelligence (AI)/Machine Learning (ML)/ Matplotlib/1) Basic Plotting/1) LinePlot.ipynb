<span style ="color:#F02F05; font-size:20px;"> <b> LINEPLOT </b> </span>

 A line plot, also known as a line chart or line graph, is a type of chart that displays data points connected by straight line segments. It's commonly used to represent trends over time or continuous data. 

# `Overview of Line Plot:`

`X-axis:` Typically represents the independent variable, which could be time, distance, or any other continuous variable.

`Y-axis: `Represents the dependent variable, which could be any measurable quantity associated with each value of the independent variable.

`Data Points:` Each data point represents a pair of values (x, y) where x is the value on the x-axis and y is the corresponding value on the y-axis.

`Line Segments:` Data points are connected by straight line segments, creating a visual representation of the relationship between the two variables.

```python
import matplotlib.pyplot as plt
```

# EX1

```python
# Sample Data
x = [1 , 2 , 3, 4, 5]  # X - Axis values
y = [2, 3, 5, 7, 11]   # Y - Axis values

# Plotting the data
plt.plot(x , y) # Create a line plot with x and y data points

# Adding labels and title
plt.xlabel('X-axis') # Add label to the x-axis
plt.ylabel('Y-axis') # Add label to the y-axis
plt.title('Simple Line Chart') # Add title to the plot 

# Displaying the plot 
plt.show() # Display the plot 
```
